#### Q.1 check the given string is palindrome or not

```python
#Palindrome
string1=input("enter string=")
rev=""
for ch in string1:
    rev=ch+rev
if string1 == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
    

```

```python
#code
l=[3,6,9,1,5,78,0,4]
sum=0
for i in l:
    sum+=i
    avg=sum/len(l)
print(f"average of List is {avg}")
```

#### Q.4 Given a tuple of intger create a tuple of all even numbers and a tuple of all odd numbers

### Q.5 Create a dictionary where:
##### keys=student name
##### values=marks(int)
### write a menu-based program where user presses a key ('A','B','C','D') Depending on the operation they want to perform on the dictionary
#####         A-Add a Student
#####         B-Update marks
#####         C-Search for a student
#####         D-Display all students and marks

```python
stud={
    "xyz":67,
    "pqr":77,
    "cid":56,
    "abc":85,
    "ijk":63
}
while True:
    print("\n---------MENU----------\n")
    print("A.Add Student\n")
    print("B.Udate Marks\n")
    print("C.Search for a Student\n")
    print("D.Display all Students and Marks\n")
    print("E.Exist\n")

    choice=input("enter the choice=").upper()

    if choice=="A":
        name=input("enter the name:")
        marks=int(input("Enter the marks:"))
        stud[name]=marks
        print("Student Added")
        
    elif choice=="B":
        name=input("enter the name:")
        marks=int(input("Enter the marks:"))
        stud[name]=marks
        print("Marks Updated")
        
    elif choice=="C":
        name=input("enter the name:")
        if name in stud:
            print(f"found studen {stud[name]}")
        else:
            print("Studen is not available in our Dictionary")
            
    elif choice=="D":
        print(stud.items())
        
    elif choice=="E":
        print("Exist")
        break
    else:
        print("Wrong Choice please remember this you are out from the program!")
        break
```

#### Q.9 Given list print alll elements that appear more than once in the list use set

```python
l1=[1,3,5,6,7,3,4,6,8,9,3,5,7,3,2,3,5]
l2=set()
l3=set()
for i in l1:
    if i in l3:
        l2.add(i)
    else:
        l3.add(i)
print(l2)
print(l3)
```

#### Q.10 Ask tje user for a string and print: (1) all unique characters (2) the count of unique characters

```python
s=input("enter the string:")
unique=" "
for ch in s:
    if ch not in unique:
        unique=unique+ch
print(f"Unique character in string={unique}")
print(f"count the unique character={len(unique)}")
```
